## 8.5 Relación entre Tuplas y Listas  
Aunque las tuplas pueden parecer similares a las listas, a menudo se usan en diferentes situaciones y para diferentes propósitos. Las tuplas son inmutables y, por lo general, contienen una secuencia heterogénea de elementos a los que se accede mediante desempaquetado o indexación. Las listas son mutables y se accede a sus elementos a través de la indexación. Los elementos no se pueden agregar, quitar o reemplazar en una tupla.  

Por ejemplo:


```python
coral_reef = ("great_barrier", "ningaloo_coast", "amazon_reef", "pickles_reef")
coral_reef[0] = "pickles_reef"
```

```python
coral_reef_list = list(coral_reef)
coral_reef_list
```

Si un elemento dentro de una tupla es mutable, puede cambiarlo. Considere la presencia de una lista como un elemento en una tupla, luego cualquier cambio en la lista se refleja en los elementos generales de la tupla. Por ejemplo:

```python
german_cars = ["porsche", "audi", "bmw"]
european_cars = ("ferrari", "volvo", "renault", german_cars)
european_cars
```

```python
european_cars[3].append("mercedes")
print(german_cars)
print(european_cars)
```

La tupla que "contiene" una lista parece cambiar cuando cambia la lista subyacente. La idea clave es que las tuplas no tienen forma de saber si los elementos dentro de ellas son mutables. Lo único que hace que un elemento sea mutable es tener un método que altere sus datos. En general, no hay manera de detectar esto. De hecho, la tupla no cambió. No podría cambiar porque no tiene métodos de mutación. Cuando la lista cambia, la tupla no recibe notificación del cambio.  

La lista no sabe si hace referencia a ella una variable, una tupla u otra lista. La tupla en sí no es mutable ya que no tiene ningún método que pueda usarse para cambiar su contenido. Del mismo modo, la cadena es inmutable porque las cadenas no tienen ningún método de mutación.

## 8.8 Uso de la función zip()  
La función zip() crea una secuencia que agrega elementos de cada uno de los iterables (pueden ser cero o más). La sintaxis de la función zip() es:

**zip(*iterables)**  

Un iterable puede ser una lista, una cadena o un diccionario. Devuelve una secuencia de tuplas, donde la i-ésima tupla contiene el i-ésimo elemento de cada uno de los iterables. La agregación de elementos se detiene cuando se agota el iterable de entrada más corto. Por ejemplo, si pasa dos iterables con uno que contiene dos elementos y otro que contiene cinco elementos, entonces la función zip() devuelve una secuencia de dos tuplas. Con un único argumento
iterable, devuelve un iterador de una tupla. Sin argumentos, devuelve un iterador vacío. Por ejemplo:



```python
# Aquí la función zip() se usa para comprimir dos iterables del tipo de lista: x e y
x = [1, 2, 3]
y = [4, 5, 6]
zipped = zip(x, y)
list(zipped)
```

Para recorrer dos o más secuencias al mismo tiempo, las entradas se pueden emparejar con la función zip(). Por ejemplo:

```python
questions = ('name', 'quest', 'favorite color')
answers = ('lancelot', 'the holy grail', 'blue')
for q, a in zip(questions, answers):
  print(f'What is your {q}? It is {a}.')
```

**Programa 8.8: Escriba una función que reciba un número variable de cadenas como argumentos. Encuentra caracteres únicos en cada cadena**

```python
def find_unique(*all_words):
  for each_word in all_words:
    unique_character_list = list(set(each_word))
    print(f"Unique characters in the word {each_word} are {unique_character_list}")

def main():
  find_unique("egg", "immune", "feed", "vacuum", "goddessship")

if __name__ == "__main__":
  main()
```

**Programa 8.8: Escriba una función que reciba un número variable de cadenas como argumentos. Encuentra caracteres únicos en cada cadena**

```python
def unique_words(user_input):
  words = user_input.split()
  print(f"The unique and sorted words are {sorted(list(set(words)))}")

def main():
  sentence = input("Enter a sentence ")
  unique_words(sentence)

if __name__ == "__main__":
  main()
```

## 8.11 Frozenset  
Un Frozenset es básicamente lo mismo que un conjunto, excepto que es inmutable. Una vez que se crea un Frozenset, sus elementos no se pueden cambiar. Dado que son inmutables, se pueden usar como miembros en otros conjuntos y como claves de diccionario. Los Frozenset tienen las mismas funciones que los conjuntos normales, excepto que ninguna de las funciones que cambian el contenido (actualizar, eliminar, abrir, etc) está disponible.

```python
dir(frozenset)
```

Lista de métodos disponibles para frozenset. Por ejemplo:
