## 7.1 Creación de diccionario

Un diccionario es una colección de un conjunto desordenado de pares clave:valor, con el requisito de que las claves sean únicas dentro de un diccionario. Los diccionarios se construyen usando llaves { }, en las que
se incluye una lista de pares clave:valor separados por comas. Además, hay dos puntos (:) que separan cada uno de estos pares de clave y valor, donde las palabras a la izquierda del operador de dos puntos son las claves y las palabras a la derecha del operador de dos puntos son los valores. A diferencia de las listas, que están indexadas por un rango de números, los diccionarios están indexados por claves. Aquí, una clave junto con su valor asociado se denomina par clave:valor . Las claves del diccionario distinguen entre mayúsculas y minúsculas.

La sintaxis para crear un diccionario es:

image.png

Por ejemplo:

```python
fish = {"g": "goldfish", "s":"shark", "n": "needlefish", "b":"barramundi", "m":"mackerel"}
fish
```

Se puede crear un diccionario vacío como se muestra en ejemplo anterior y el empty_dictionary es del tipo dict. En los diccionarios, el orden de los pares **clave:valor** no importa. Por ejemplo:

```python
pizza = {"pepperoni":3, "calzone":5, "margherita":4}
fav_pizza = {"margherita":4, "pepperoni":3, "calzone":5}
pizza == fav_pizza
```

Los diccionarios pizza y fav_pizza tienen los mismos pares **clave:valor** pero en un orden diferente. Si compara estos dos diccionarios, da como resultado un valor booleano verdadero. Esto indica que el orden de los pares **clave:valor** no importa en los diccionarios. No se permite trocear en los diccionarios ya que no están ordenados como listas.

**IMPORTANTE:** Antes de la versión Python 3.6, la ejecución de declaraciones de diccionario generaba una salida desordenada de pares clave:valor . Sin embargo, a partir de la versión Python 3.6, la salida de las sentencias del diccionario se ordena en pares clave:valor . Aquí, ordenado significa “inserción ordenada”, es decir, los diccionarios recuerdan el orden en que se insertaron los pares clave:valor.

## 7.2 Acceso y modificación de pares clave:valor en diccionarios
Se puede acceder a cada par clave:valor individual en un diccionario a través de claves especificándolo entre corchetes. La clave proporcionada entre corchetes indica la **clave:valor** al que se accede

La sintaxis para acceder al valor de una clave en el diccionario es:

**nombre_diccionario[clave]**

La sintaxis para modificar el valor de una clave existente o para agregar un nuevo par clave:valor a un diccionario es:

**nombre_diccionario[clave] = valor**

Si la clave ya está presente en el diccionario, entonces la clave se actualiza con el nuevo valor. Si la clave no está presente, el nuevo par clave:valor se agrega al diccionario. Por ejemplo:

### 7.2.1 La función dict()
La función dict() incorporada se usa para crear un diccionario. La sintaxis de la función dict() cuando los argumentos de palabra clave opcionales utilizados es:

**dict([\*\*kwarg])**

La función dict() devuelve un nuevo diccionario inicializado a partir de argumentos de palabras clave opcionales y un conjunto posiblemente vacío de argumentos de palabras clave. Si no se proporciona ningún argumento de
palabra clave, se crea un diccionario vacío. Si se proporcionan argumentos de palabra clave, los argumentos de palabra clave y sus valores de la forma **kwarg = valor** se agregan al diccionario como pares clave:valor. Por ejemplo:

```python
numbers = dict(one=1, two=2, three=3)
numbers
```

```python
numbers2 = {"one":1, "two":2, "three":3}
numbers2
```

La sintaxis para la función dict() cuando se usan iterables es:  

**dict(iterable[\*\*kwarg])**  

Puede especificar un iterable que contenga exactamente dos objetos como tupla, la clave y el valor en la función dict(). Por ejemplo:


```python
dict([('sape', 4139), ('guido', 4127), ('jota', 4098)])
```

La función dict() crea diccionarios directamente a partir de secuencias de pares de tuplas de clave y valor.

### 7.4.1 Llenar diccionarios con pares clave:valor

**Programa 7.1:** Programa para construir dinámicamente la entrada del usuario como una lista

```python
def main():
  tot_items = int(input("Enter number items"))
  tot_items += 1

  print("Method 1: Building Dictionaries")
  
  build_dictionary = {}
  for i in range(0, tot_items):
    dic_key = input("Enter key ")
    dic_val = input("Enter val ")
    build_dictionary.update({dic_key: dic_val})

  print(f"Dictionary is {build_dictionary}")
  
  print("Method 2: Building Dictionaries")

  build_dictionary = {}
  for i in range(0, tot_items):
    dic_key = input("Enter key ")
    dic_val = input("Enter val ")
    build_dictionary[dic_key] = dic_val

  print(f"Dictionary is {build_dictionary}")
  print("Method 3: Building Dictionaries")

  build_dictionary = {}
  i = 0
  while i <= tot_items:
    dict_key = input("Enter key ")
    dict_val = input("Enter val ")
    build_dictionary.update({dict_key: dict_val})
    i = i + 1

  print(f"Dictionary is {build_dictionary}")
  
if __name__ == "__main__":
  main()
```

### 7.4.2 Travesía del Diccionario
Se puede usar un bucle for para iterar sobre claves o valores o pares clave:valor en diccionarios. Si itera sobre un diccionario usando un bucle for , entonces, de manera predeterminada, iterará sobre las claves. Si desea
iterar sobre los valores, use el método de values() y para iterar sobre los pares clave:valor , especifique explícitamente el método de items() del diccionario . Los tipos de datos dict_keys, dict_values y dict_items devueltos por métodos de diccionario se pueden usar en bucles for para iterar sobre las claves o valores o pares clave:valor.


**Programa 7.2:** Programa para ilustrar el desplazamiento de pares clave:valor en diccionarios utilizando el ciclo for

```python
currency = {"India": "Rupee", "USA": "Dollar", "Russia": "Ruble", "Japan": "Yen", "Germany": "Euro"}

def main():
  print("List of Countries")
  for key in currency.keys():
    print(key)
  
  print("List of Currencies in different Countries")
  for value in currency.values():
    print(value)

  for key, value in currency.items():
    print(f"'{key}' has a currency of type '{value}'")

if __name__ == "__main__":
  main()
```

**Programa 7.3:** Escribir un programa en Python para comprobar la presencia de una clave en el diccionario y sumar todos sus valores

```python
historical_events = {"apollo11": 1969, "great_depression": 1929, "american_revolution":1775, "berlin_wall": 1989}

def check_key_presence():
  key = input("Enter the key to check for its presence ")
  if key in historical_events.keys():
    print(f"Key '{key}' is present in the dictionary")
  else:
    print(f"Key '{key}' is not present in the dictionary")

def sum_dictionary_values():
  print("Sum of all the values in the dictionary is")
  print(f"{sum(historical_events.values())}")

def main():
  check_key_presence()
  sum_dictionary_values()

if __name__ == "__main__":
  main()
```
