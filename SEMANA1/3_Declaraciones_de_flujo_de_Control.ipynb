Python admite un conjunto de declaraciones de flujo de control que puede integrar en su programa. Las declaraciones dentro de su programa Python generalmente se ejecutan secuencialmente de arriba a abajo, en el orden en que aparecen. Además de las declaraciones de flujo de control secuencial, puede emplear declaraciones de flujo de control de toma de decisiones y bucles para dividir el flujo de ejecución, lo que permite que su programa ejecute condicionalmente bloques de código particulares.

El término flujo de control detalla la dirección que toma el programa.
Las declaraciones de flujo de control (FIGURA) en el lenguaje de programación Python son:

1. **Sentencias de Flujo de Control Secuencial:** Se refiere a la ejecución línea por línea, en la que las sentencias se ejecutan secuencialmente, en el mismo orden en que aparecen en el programa. 
2. **Sentencias de Flujo de Control de Decisión:** Dependiendo de si una condición es Verdadera o Falsa, la estructura de decisión puede omitir la ejecución de un bloque completo de sentencias o incluso ejecutar un bloque de sentencias en lugar de otro (if, if…else y if…elif …más).
3. **Sentencias de flujo de control de bucle:** esta es una estructura de control que permite la ejecución de un bloque de sentencias varias veces hasta que se cumple una condición de terminación de bucle (bucle for y bucle while ). Las declaraciones de flujo de control de bucle también se denominan declaraciones de repetición o declaraciones de iteración.

## Sentencia if  
La sentencia **if EXPRESION:**, significa, Si se cumple la expresión condicional se ejecuta el bloque de sentencias seguidas.

```python
monto = 15
if monto > 5:
  monto += 5

print(monto)
```

```python
monto = 15
if monto > 5:
  pass

print(monto)
```

**Expresión ==**

```python
monto_1 = 7
monto_2 = 7
if monto_1 == monto_2:
    monto_1 += monto_2

print(monto_1)
```

```python
# Error al usar el operador de asignación como operador de comparación
monto_1 = 7
monto_2 = 7
if monto_1 = monto_2:
    monto_1 += monto_2

print(monto_1)
```

## Sentencia elif  
La sentencia elif EXPRESION, significa, De lo contrario Si se cumple la expresión condicional se ejecuta el bloque de sentencias seguidas.

```python
monto_1 = 4
monto_2 = 7
if monto_1 == monto_2:
  monto_1 += monto_2
elif monto_1 > monto_2:
  monto_1 *= monto_2
elif monto_1 < monto_2:
  monto_1 += 10

print(monto_1)
```

## Sentencia else
La sentencia else, significa, De lo contrario se cumple sin evaluar ninguna expresión condicional y ejecuta el bloque de sentencias seguidas.

```python
monto_1 = 9
monto_2 = 5
if monto_1 == monto_2:
  monto_1 += monto_2
else:
  monto_1 *= monto_2

print(monto_1)
```

```python
1 is 1
```

```python
a = 10
b = 10
a is b
```

```python
a = 10
b = 20
a is b
```

```python
a = [1, 2, 3]
b = [1, 2, 3]
a == b
```

```python
a = [1, 2, 3]
b = [1, 2, 3]
a is b
```

```python
a = [1, 2, 3]
b = a
a is b
```

**Conclusión:** el operador **is** devuelve verdadero cuando los dos objetos a comparar son el mismo. Por otro lado, el operador **==** devuelve verdadero si el contenido es el mismo, pero no necesariamente el objeto. Cuando hablamos de tipos de datos básicos. como los enteros, reales o cadenas de texto el comportamiento de los dos operadores **==** e **is** es similar.

## Operador in
El operador in, significa, para cualquier colección del valor del lado izquierdo contenga el valor del lado derecho:

```python
Lista = [1, 2, 3, 4, 5, 6, (1,2,3), (4,5,6)]
print(2 in Lista)
print(3 in Lista)
print((1,2,3) in Lista)
print((4,5,6) in Lista)
print(12 in Lista)
```

## Operador not in
El operador not in, el contrario de operador in, devuelve True cuando un elemento no está en una secuencia.

```python
Lista = ["a", "b", 3, 6, (1,2,3)]
print(4 not in Lista)
print("a" not in Lista)
print((1,2,3) not in Lista)
```

## Bucle while
En Python tiene una palabra reservada llamada while que nos permite ejecutar ciclos, o bien secuencias periódicas que nos permiten ejecutar código múltiples veces. El ciclo while nos permite realizar múltiples iteraciones basándonos en el resultado de una expresión lógica que puede tener como resultado un valor True o False.

**Bucle “while” controlado por Conteo**

```python
numero = 0
suma = 0
while numero <= 10:
    numero += 1
    suma += numero
    print(numero, ",", suma)

print("La suma es: " + str(suma))
```

**Bucle “while” controlado por Evento**  

```python
total = 0
contar = 0
nota = int(input("Introduzca la nota de un estudiante (-1 para salir): "))
while nota != -1:
    total += nota
    contar += 1
    nota = int(input("Introduzca la nota de un estudiante (-1 para salir): "))
   
promedio = total / contar
print("El acumulado de notas es: ", total)
print("El conteo es: ", contar)
print ("Promedio de notas del grado escolar es: " + str(promedio))
```

**Bucle “while” con “else”**  
Al igual que la sentencia if, la estructura while también puede combinarse con una sentencia else).

El nombre de la sentencia else es equivocada, ya que el bloque else se ejecutará en todos los casos, es decir, cuando la expresión condicional del while sea False, (a comparación de la sentencia if).

```python
total = 0
contar = 0
nota = int(input("Introduzca la nota de un estudiante (-1 para salir): "))
while nota != -1:
    total += nota
    contar += 1
    nota = int(input("Introduzca la nota de un estudiante (-1 para salir): "))
else:
    promedio = total / contar
    print ("Promedio de notas del grado escolar: " + str(promedio))
```

## Sentencias utilitarias

**Sentencia break**  
A continuación, se presenta un ejemplo del uso del bucle while controlado la sentencia break:

```python
variable = 10
while variable > 0:
    print ("Actual valor de variable:", variable)
    variable -= 1
    if variable == 5:
        break
```

**Sentencia continue**  
A continuación, se presenta un ejemplo del uso del bucle while controlado la sentencia continue:

```python
variable = 10
while variable > 0:              
   variable = variable -1
   if variable == 5:
      continue
      
   print ("Actual valor de variable:", variable)
```

La sentencia continue hace que pase de nuevo al principio del bucle aunque no se haya terminado de ejecutar el ciclo anterior.

## Bucle for  
La sentencia for en Python difiere un poco de lo que uno puede estar acostumbrado en lenguajes como C o Pascal. En lugar de siempre iterar sobre una progresión aritmética de números (como en Pascal) o darle al usuario la posibilidad de definir tanto el paso de la iteración como la condición de fin (como en C), la sentencia for de Python itera sobre los ítems de cualquier secuencia (una lista o una cadenas de caracteres), en el orden que aparecen en la secuencia.

```python
for i in range(10):
  print(i)
```

```python
for i in range(0,10):
  print(i)
```

```python
for i in range(1,10):
  print(i)
```

```python
for i in range(0,10,2):
  print(i)
```

```python
cadena = "cadena de texto"

for valor in cadena:
  print(valor)
```

**Bucle “for” con cadenas de texto y función “range”**

```python
# En el código debemos identificar lo siguiente:
# El argumento: "\n" es un salto de línea
# En el ciclo for in range, la variable palabra almacena
# las posiciones de las palabras, la primera es cero
# La función len(frase) nos da la longitud total de la frase
# El bucle for in range recorre 38 posiciones
# empezando por el cero, lo que nos da un rango (0, 37)
# El método format dentro del argumento de la función print()
# concatena cadenas de texto de modo correspondiente

frase = "Entre broma y broma la verdad se asoma"
print("La oración analizada es:", frase)
print("la longitud de la frase es:", len(frase), "\n")
for letra in range(len(frase)):
    print ("letra: {0}, en la frase su posición es: {1}".format(
        frase[letra], letra))
```

Analisis del ciclo for con la función range

```python
# Si colocamos un solo argumento: N
# el ciclo empezará por cero hasta valor del argumento menos 1: N-1
# Se cunple que se recorre N elementos
for item in range(10):
  print(item)
```

```python
# Si colocamos un dos argumentos: N1, N2
# el ciclo empezará por el primer elemento N1 hasta el último argumento menos 1: N2-1
# Se cumple que se recorre N2-N1 elementos
for item in range(7, 10):
  print(item)
```

```python
# Caso de tres argumentos
for item in range(1, 10, 2):
  print(item)
```

```python
# Caso de tres argumentos
for item in range(1, 11, 2):
  print(item)
```
