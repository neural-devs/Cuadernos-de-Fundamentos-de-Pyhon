## 9.1 Tipos de Archivos  
Python admite dos tipos de archivos: archivos de texto y archivos binarios. Estos dos tipos de archivos pueden parecer iguales en la superficie, pero codifican los datos de manera diferente. Si bien los archivos binarios y de texto contienen datos almacenados como una serie de bits (valores binarios de 1 y 0), los bits de los archivos de texto representan caracteres, mientras que los bits de los archivos binarios representan datos personalizados.

Los archivos binarios suelen contener una secuencia de bytes o agrupaciones ordenadas de ocho bits. Al crear un formato de archivo personalizado para un programa, un desarrollador organiza estos bytes en un formato que almacena la información necesaria para la aplicación. Los formatos de archivo binario pueden incluir múltiples tipos de datos en el mismo archivo, como datos de imagen, video y audio. Estos datos pueden ser interpretados por programas compatibles, pero aparecerán como texto ilegible en un editor de texto. Cuando la imagen se abre en un editor de texto, los datos binarios se convierten en texto irreconocible. Sin embargo, puede notar que parte del texto es legible. Esto se debe a que el formato JPG incluye pequeñas secciones para
almacenar datos de texto. El editor de texto, aunque no está diseñado para leer este formato de archivo, todavía muestra este texto cuando se abre el archivo. Muchos otros tipos de archivos binarios también incluyen secciones de texto legible. Por lo tanto, es posible encontrar alguna información sobre un tipo de archivo binario desconocido abriéndolo en un editor de texto. Los archivos binarios a menudo contienen encabezados, que son bytes de datos al principio de un archivo que identifica el contenido del archivo. Los encabezados suelen incluir el tipo de archivo y otra información descriptiva. Si un archivo tiene una información de encabezado no válido, es posible que un programa de software no abra el archivo o informe que el archivo está dañado.



Los archivos de texto son más restrictivos que los archivos binarios, ya que solo pueden contener datos de texto. Sin embargo, a diferencia de los archivos binarios, es menos probable que se dañen. Mientras que un pequeño error en un archivo binario puede hacer que sea ilegible, un pequeño error en un archivo de texto simplemente puede aparecer una vez que se ha abierto el archivo. Un archivo de texto sin formato típico contiene varias líneas de texto, cada una seguida de un carácter de fin de línea (EOL). Se coloca un marcador de fin de archivo (EOF) después del carácter final, que señala el final del archivo. Los archivos de texto incluyen un esquema de codificación de caracteres que determina cómo se interpretan los caracteres y qué caracteres se pueden mostrar. Dado que los archivos de texto utilizan un formato estándar simple, muchos programas pueden leer y editar archivos de texto. Los editores de texto comunes incluyen Microsoft Notepad y WordPad, que se incluyen con Windows, y Apple TextEdit, que se incluye con Mac OS X.  

Por lo general, podemos saber si un archivo es binario o de texto en función de su extensión de archivo. Esto se debe a que, por convención, la extensión refleja el formato del archivo y, en última instancia, es el formato del
archivo el que dicta si los datos del archivo son binarios o de texto.

### 9.1.1 Rutas de archivos  
Todos los sistemas operativos siguen las mismas convenciones generales de nomenclatura para un archivo individual: un nombre de archivo base y una extensión opcional, separados por un punto. Tenga en cuenta que un directorio es simplemente un archivo con un atributo especial que lo designa como directorio, pero por lo demás debe seguir las mismas reglas de nomenclatura que un archivo normal. Para hacer uso de archivos, debe proporcionar una ruta de archivo, que es básicamente una ruta para que el usuario o el programa sepa dónde se encuentra el archivo. La ruta a un archivo específico consta de uno o más componentes, separados por un carácter especial (una barra invertida para Windows y una barra inclinada para Linux), y cada componente suele ser un
nombre de directorio o de archivo, y posiblemente un nombre de volumen o de unidad. en Windows o root en Linux. Si un componente de una ruta es un nombre de archivo, debe ser el último componente. A menudo es crítico para la interpretación del sistema de un camino lo que el principio, o prefijo, de la ruta parece. En el Sistema Operativo Windows, la longitud máxima de una ruta es de 260 caracteres y en el Sistema Operativo Linux la longitud máxima de la ruta es de 4096 caracteres.


Las siguientes reglas fundamentales permiten que las aplicaciones creen y procesen nombres válidos para archivos y directorios en los sistemas operativos Windows y Linux, a menos que se especifique explícitamente:  

- Use un punto para separar el nombre del archivo base de la extensión en el nombre del archivo.
- En Windows, use la barra invertida (\\) y en Linux use la barra inclinada (/) para separar los componentes de una ruta. La barra invertida (o barra diagonal) separa un nombre de directorio de otro nombre de directorio en una ruta y también divide el nombre de archivo de la ruta que conduce a él. La barra invertida (\\) y la barra inclinada (/) son caracteres reservados y no puede utilizarlos en el nombre del archivo o directorio real.
- No asuma la distinción entre mayúsculas y minúsculas. Los nombres de archivos y directorios en Windows no distinguen entre mayúsculas y minúsculas, mientras que en Linux sí lo hacen. Por ejemplo, los nombres de directorio ORANGE, Orange y orange son los mismos en Windows pero son diferentes en el sistema operativo
Linux.
- En Windows, los designadores de volumen (letras de unidad) no distinguen entre mayúsculas y minúsculas. Por ejemplo, "D:\" y "d:\" se refieren a la misma unidad.
- Los caracteres reservados que no deben utilizarse para nombrar archivos y directorios son < (menor que), > (mayor que),: (dos puntos), " (comillas dobles), / (barra diagonal), \ (barra diagonal inversa), | (barra vertical o 
tubería), ? (signo de interrogación) y * (asterisco).
- En el sistema operativo Windows, palabras reservadas como CON, PRN, AUX, NUL, COM1, COM2, COM3, COM4, COM5, COM6, COM7, COM8, COM9, LPT1, LPT2, LPT3, LPT4, LPT5, LPT6, LPT7, LPT8, y LPT9 no debe usarse para nombrar archivos y directorios.



### 9.1.2 Ruta completamente calificada y ruta relativa  
Una ruta de archivo puede ser un nombre de ruta completo o una ruta relativa. El nombre de vía de acceso completo también se denomina vía de acceso absoluta. Se dice que una ruta es una ruta completa si apunta a la ubicación del archivo, que siempre contiene la raíz y la lista completa de directorios. El directorio actual es el directorio en el que un usuario está trabajando en un momento dado. Cada usuario siempre está trabajando dentro de un directorio.  

Tenga en cuenta que el directorio actual puede o no ser el directorio raíz dependiendo de cómo se configuró durante la operación de "cambio de directorio" más reciente en ese disco. 

El directorio raíz, a veces simplemente llamado raíz, es el directorio "más alto" en la jerarquía. También puede pensar en él en general como el comienzo o el comienzo de una estructura de directorio particular. El directorio raíz contiene todos los demás directorios de la unidad y, por supuesto, también puede contener archivos. Por ejemplo, el directorio raíz de la partición principal en su sistema Windows será C:\ y el directorio raíz en su sistema Linux será / (barra inclinada).

### 9.2.2 El Método close()  
Abrir archivos consume recursos del sistema y, según el modo de archivo, es posible que otros programas no puedan acceder a ellos. Es importante cerrar el archivo una vez finalizado el procesamiento. Una vez que se cierra el objeto del controlador de archivos, ya no puede leer ni escribir en el archivo. Cualquier intento de utilizar el objeto del controlador de archivos después de cerrarlo dará como resultado un error.  

La sintaxis de la función close() es:

**manejador_de_archivos.close()**

Por ejemplo:


```python
manejador_archivo = open("ejemplo.txt","r")
manejador_archivo.close()
```

Debe llamar a **manejador_archivo.close()** para cerrar el archivo. Esto libera inmediatamente cualquier recurso del sistema utilizado por él ÿ. Si no cierra explícitamente un archivo, el recolector de basura de Python eventualmente destruirá el objeto y cerrará el archivo abierto por usted, pero es posible que el archivo haya permanecido abierto por un tiempo. Otro riesgo es que diferentes implementaciones de Python harán esta limpieza en diferentes momentos.  

Si se produce una excepción al realizar alguna operación en el archivo, el código sale sin cerrar el archivo. Para superar este problema, debe usar un bloque try-except finalmente para manejar las excepciones. Por ejemplo:  

```python
try:
  f = open("file", "w")
  try:
    f.write('Hello World!')
  finally:
    f.close()
except IOError:
  print('oops!')
```

```python
try:
  f = open("ejemplo.txt", "r")
  try:
    f.write('Hello World!')
  finally:
    f.close()
except IOError:
  print('oops!')
```

### 9.2.3 Uso de declaraciones with para abrir y cerrar archivos

En lugar de usar bloques try-except-finally para manejar las operaciones de apertura y cierre de archivos, una forma mucho más limpia de hacer esto en Python es usar la instrucción with . Puede usar una declaración with en Python de modo que no tenga que cerrar el objeto del controlador de archivos. La civilización se fusionó alrededor del 3150 a. En la sintaxis, las palabras with y as son palabras clave y la palabra clave with va seguida de la función open() y termina con dos puntos. La palabra clave as actúa como un alias y se usa para asignar el objeto que regresa de la función open() a una nueva variable file_handler. 

La declaración crea un administrador de contexto y cerrará automáticamente el objeto del controlador de archivos cuando haya terminado con él, incluso si se genera una excepción en el camino y, por lo tanto, administrará correctamente los recursos.

**Programa 9.2:** Programa para leer e imprimir cada línea en el archivo "egipto.txt" usando la declaración . El contenido de muestra del archivo "japan.txt" se proporciona a  continuación.

```python
def read_file():
  print("imprimiendo cada línea del archivo:")
  print("----------------------------------")

  with open("egipto.txt") as file_handler:
    for each_line in file_handler:
      print(each_line, end="")

def main():
  read_file()

if __name__ == "__main__":
  main()
```

**Programa 9.3:** Escriba un programa Python para leer el archivo "roma.txt" utilizando el método read(). El  contenido de muestra del archivo "roma.txt" se muestra a continuación:  

***El Imperio es el sistema en el que el poder político real estaba en manos de un solo individuo, el emperador. En Roma, se inauguró con el emperador Augusto. El Senado quedó limitado a ser un órgano de apoyo de ese poder político.***

```python
def main():
  with open("roma.txt") as file_handler:
    print("Imprimir todo el contenido del archivo:")
    print("---------------------------------------")
    print(file_handler.read(), end=" ")

if __name__ == "__main__":
  main()
```

**Programa 9.4:** Considere el archivo "roma.txt" especificado en el Programa 9.3. Escriba el programa Python para leer el archivo "rome.txt" usando el método readline()

```python
def main():
  with open("roma.txt") as file_handler:
    print("Imprimir una sola línea del archivo:")
    print("-----------------------------------")
    print(file_handler.readline(), end="")
    print("Imprime otra sola línea del archivo:")
    print("-----------------------------------")
    print(file_handler.readline(), end="")

if __name__ == "__main__":
  main()
```

**Programa 9.5:** Considere el archivo "roma.txt" especificado en el Programa 9.3. Escriba el
programa Python para leer el archivo "rome.txt" usando el método readlines()

```python
def main():
  with open("roma.txt") as file_handler:
    print("Imprime el contenido del archivo como una lista")
    print(file_handler.readlines())

if __name__ == "__main__":
  main()
```

**El método readline()** devuelve una lista de cadenas y cada línea es un elemento de lista. Se deja un carácter de nueva línea (\n) al  final de cada elemento de cadena de la lista que indica que se ha llegado al final de la línea. Solo se omite en la última línea del archivo si el archivo no termina en un carácter de nueva línea.

**El método seek()** se utiliza para establecer la posición actual del controlador de archivos. Nunca olvide que cuando
administre archivos, siempre habrá una posición dentro de ese archivo en la que esté trabajando actualmente. Cuando
abre un archivo, esa posición es el comienzo del archivo, pero a medida que trabaja con él, puede avanzar. El método
seek() le será útil cuando necesite trabajar con ese archivo abierto.  

Por ejemplo:

```python
f = open('workfile', 'w')
f.write('0123456789abcdef')
f.close()
```

```python
f = open('workfile', 'r')
f.seek(5)
f.read()
```

```python
f.seek(1)
f.read()
```
